```python
import soccerdata as sd
import pandas as pd
import numpy as np
print("Todo instalado correctamente ✓")
```

```python
fbref = sd.FBref(
    leagues=[
        "Big 5 European Leagues Combined",
        "INT-World Cup",
        "INT-European Championship"
    ], 
    seasons=2025
)

stats = fbref.read_player_season_stats(stat_type="standard")

print(f"Datos obtenidos: {stats.shape[0]} registros ✓")
print(stats.head())
```

```python
print(stats.columns.tolist())
print(f"\nColumnas disponibles: {len(stats.columns)}")
```

```python
# Resetear índice y eliminar columnas extra
stats = stats.reset_index()
stats = stats.drop(columns=['level_0', 'index'])

# Limpiar age manejando valores nulos
stats['age'] = stats['age'].str.split('-').str[0]
stats['age'] = pd.to_numeric(stats['age'], errors='coerce')
stats['age'] = stats['age'].fillna(0).astype(int)

print(f"Dataset limpio: {stats.shape[0]} jugadores ✓")
print(stats.head())
```

```python
# Guardar dataset de FBref
stats.to_csv(r"C:\Users\mrabe\OneDrive\Documentos\Scouting Dashboard\Data\fbref_stats.csv", index=False)
print("Dataset FBref guardado ✓")
```

```python
players_tm = pd.read_csv(r"C:\Users\mrabe\OneDrive\Documentos\Scouting Dashboard\data\players.csv")

players_tm = players_tm[players_tm['last_season'] >= 2025]

players_tm = players_tm[[
    'name', 'position', 'sub_position', 'date_of_birth',
    'current_club_name', 'current_club_domestic_competition_id',
    'market_value_in_eur', 'highest_market_value_in_eur',
    'foot', 'height_in_cm', 'contract_expiration_date',
    'country_of_citizenship', 'last_season'
]]

players_tm = players_tm[players_tm['market_value_in_eur'].notna()]

# Calcular edad
players_tm['date_of_birth'] = pd.to_datetime(players_tm['date_of_birth'])
players_tm['edad'] = 2025 - players_tm['date_of_birth'].dt.year

print(f"Jugadores activos con valor de mercado: {players_tm.shape[0]} ✓")
print(players_tm.head())
```

```python
from unidecode import unidecode

# Normalizar nombres en ambos datasets
stats['player_norm'] = stats['player'].apply(unidecode).str.lower().str.strip()
players_tm['name_norm'] = players_tm['name'].apply(unidecode).str.lower().str.strip()

# Cruzar con nombres normalizados
merged = pd.merge(stats, players_tm, left_on='player_norm', right_on='name_norm', how='inner')

print(f"Jugadores que coinciden ahora: {merged.shape[0]}")
print(f"Jugadores perdidos: {stats.shape[0] - merged.shape[0]}")
```

```python
# Borrar columnas redundantes
merged = merged.drop(columns=['current_club_domestic_competition_id', 'pos', 'born', 'season'])

# Guardar dataset final
merged.to_csv(r"C:\Users\mrabe\OneDrive\Documentos\Scouting Dashboard\data\players_final.csv", index=False)

print(f"Dataset final guardado: {merged.shape[0]} jugadores ✓")
print(merged.columns.tolist())
```
